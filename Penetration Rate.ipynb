```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
from scipy import stats

%matplotlib inline
```

```python
base_size  =  110
name = 'result_1944_1.000000_15_8m250.csv'
folder = '2002DB1a'
df = pd.read_csv('../data-output/output_2002_DB1a/{}'.format(name))

rank_points = [0.01*base_size,0.05*base_size,0.1*base_size,0.15*base_size,0.2*base_size,0.25*base_size,0.3*base_size,0.35*base_size,0.4*base_size]
quality_points = [1,2,3,4,5]
color=['gray','blue','green','cyan','red','black']
pos = []
pos_color = 0
plt.figure(figsize=(15,15))
value_per_quality = []


for quality in quality_points:
    
    df_quality_rank = df[(df['quality']==quality) ]
    value_per_quality.append(df_quality_rank.shape[0])
    if(df_quality_rank.shape[0] > 0):

        for rank in rank_points:
            df_rank = df_quality_rank[(df_quality_rank['pos_rank'] <=rank) & (df_quality_rank['pos_rank'] > 0)]
            error = 1 - df_rank.shape[0]/df_quality_rank.shape[0]
            pos.append([rank/base_size*100,error])
        pos = np.array(pos)
        plt.plot(pos[:,0],pos[:,1],marker='x',color=color[pos_color])
    pos = []
    pos_color += 1    


for rank in rank_points:
    df_rank = df[(df['pos_rank'] <=rank) & (df['pos_rank'] > 0)]
    error = 1 - df_rank.shape[0]/df.shape[0]
    pos.append([rank/base_size*100,error])
pos = np.array(pos)
plt.plot(pos[:,0],pos[:,1],marker='x',color=color[pos_color])
    
        
plt.grid(True) 
plt.xlabel('Base Penetration (%)')
plt.ylabel('Error Rate (%)')
plt.title('Penetration Rate {}'.format(name))
#plt.xlim(xmin=0,xmax=2010)
gray_patch = mpatches.Patch(color='gray', label="quality 1")
blue_patch = mpatches.Patch(color='blue', label="quality 2")
green_patch = mpatches.Patch(color='green', label="quality 3")
cyan_patch = mpatches.Patch(color='cyan', label="quality 4")
red_patch = mpatches.Patch(color='red', label="quality 5")
black_patch = mpatches.Patch(color='black', label="quality total")


plt.legend(handles=[gray_patch,blue_patch,green_patch,black_patch])
#plt.legend("","quality 25-50","quality 50-75","quality 75-100")

plt.legend(handles=[gray_patch,blue_patch,green_patch,cyan_patch,red_patch,black_patch])

plt.savefig('{}/penetration_rate_{}.png'.format(folder,name))
plt.show();
plt.clf();
```
